#### Ejercicio 01. ¿Cuál es la casa más cara en todo el dataset? (★☆☆)

Imprime la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle del Prado, Nº20 es la más cara y su precio es de 5000000 USD`

```python
# TODO

################

precio_mas_cara = ds['price'].max()

direccion_mas_cara = ds.loc[ds['price']==precio_mas_cara, 'address'].iloc[0]
level_3_mas_cara = ds.loc[ds['price']==precio_mas_cara, 'level3'].iloc[0]
level_7_mas_cara = ds.loc[ds['price']==precio_mas_cara, 'level7'].iloc[0]
level_1_mas_cara = ds.loc[ds['price']==precio_mas_cara, 'level1'].iloc[0]
country_mas_cara = ds.loc[ds['price']==precio_mas_cara, 'country'].iloc[0]

print(f'El precio de la casa mas cara es {precio_mas_cara} Euros, cuya direccion corresponde a {direccion_mas_cara}, {level_3_mas_cara}, {level_7_mas_cara}, {level_1_mas_cara}, {country_mas_cara}.')
```

#### Ejercicio 02. ¿Cuál es la casa más barata del dataset? (★☆☆)

Este ejercicio es similar al ejercicio anterior, solo que ahora buscamos la casa con el precio más bajo. Recuerda imprimir la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle Alcalá, Nº58 es la más barata y su precio es de 12000 USD`

```python
# TODO

precio_mas_barato = ds['price'].min()
#print(precio_mas_barato)    # Precio = 0 --> Dato NO REAL

casas_mayor_0 = ds.loc[ds['price']>precio_mas_barato]        # Filtro los mayores a cero

ordenado = casas_mayor_0.sort_values('price', ascending = True)         # Ordeno para verificar
#print(ordenado)                                                        # Nuevo precio minimo = 600


# USANDO .min()

minimo_nuevo = ordenado['price'].min()
direccion_minimo = ordenado.loc[ordenado['price']==minimo_nuevo, 'address'].iloc[0]

print(f'La casa mas barata tiene un precio de {minimo_nuevo} Euros, y su direccion es {direccion_minimo}.')
```

#### Ejercicio 03. ¿Cuál es la casa más grande y la más pequeña del dataset? (★☆☆)

Imprime la dirección y el área de las casas seleccionadas. Por ejemplo:

Este ejercicio es similar al anterior, pero buscamos las casas más grandes y pequeñas en base a su superficie.

#### Ejercicio 05. ¿El dataset contiene valores nulos (NAs)? (★☆☆)

Imprima un booleano (`True` o `False`) para verificar si hay valores nulos seguido de las columnas que contiene el NAs.

```python
# TODO

dn = ds.isnull().sum().reset_index(drop=False)          # Uso reset_index para colocar los indices (nombres de las columnas) como nueva columna

for i in range(37):                      # 37 = Numero de columnas
    if dn.iloc[i,1] > 0:
        print(True, dn.iloc[i,0])
    else:
        print(False, dn.iloc[i,0])
```

#### Ejercicio 06. Elimina los valores nulos (NAs) del dataset, si aplica (★★☆)

Despues de eliminar los valores nulos, compara el tamaño del DataFrame antes y después de la eliminación.


```python
# TODO

#print(ds.isnull().sum())                                   # Reviso cantidad de valores nulos por columna

ds2 = ds.dropna(axis=1, how='all')                          # Elimino las columnas zipcode y customzone: esta COMPLETAMENTE LLENAS DE NaN.
ds3 = ds2.drop(columns=['level4','level6','level8'])        # Elimino columnas level4, level6 y level8, ya que poseen MUCHOS valores NaN, y no son muy importantes.

#print(ds3.info())                                          # Reviso el TIPO de cada columna

dicci = {'realEstate_name':'No especificado',
         'phone_realEstate':0,
         'rooms':ds3['rooms'].mean(),
         'bathrooms':ds3['bathrooms'].mean(),
         'surface':ds3['surface'].mean(),
         'description':'No especificado',
         'level7':'No especificado'}                        # Creo diccionario --> columna:relleno

ds4 = ds3.fillna(dicci)                                     # Relleno NaN usando el diccionario (rooms, bathrooms y surface los relleno con el promedio; puede haber afectacion del analisis posteriormente.)

#print(ds4.groupby('realEstate_name').get_group('No especificado'))         # Revisando que funcionara bien el relleno

#print(ds4.isnull().sum())                                  # Compruebo que todas las columnas esten sin NaN

print(f'Se eliminaron los valores nulos. El dataframe original poseia las dimensiones {ds.shape}, mientras que el dataframe nuevo posee las dimensiones {ds4.shape}.')
```

#### Ejercicio 08. Trazar el histograma de los precios para la población de "Arroyomolinos (Madrid)" y explica qué observas (★★☆)

Imprime el histograma de los precios y escribe en la celda del Markdown un breve análisis del trazado.


```python
# TODO:

import matplotlib.pyplot as plt

filtro = da['level5'] == 'Arroyomolinos (Madrid)'
precios = da[filtro]['price']
precios = precios.reset_index(drop=True).sort_values(ascending=False)            
#print(precios,'\n')                                                          # 137 casas. 2 de ellas con precio = 0 (probablemente erroneo); se dejan ahi

#print(precios.value_counts())                                                # Hay 82 valores unicos de precios

plt.hist(precios,bins=70)                                                     # Pareciera una distribucion normal
plt.show()
```
